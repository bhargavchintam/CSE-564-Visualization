```python
import pandas as pd
import numpy as np

# Function to replace 'Unknown' with a random value from the same column
def replace_unknown_with_random(column):
    known_values = column[column != 'Unknown']  # Filter out the 'Unknown' values
    if not known_values.empty:
        # Replace 'Unknown' values with a random choice from known values
        column[column == 'Unknown'] = np.random.choice(known_values, size=sum(column == 'Unknown'))
    return column

# Apply the function to each specified column
columns_to_replace = ['Referee', 'Assistant 1', 'Assistant 2']
for col in columns_to_replace:
    df[col] = replace_unknown_with_random(df[col])

df
```

```python
# Function to select only the last two words of a string
def select_last_two_words(s):
    words = s.split()  # Split the string into a list of words
    return ' '.join(words[-2:]) if len(words) >= 2 else s

# Apply the function to each specified column
columns_to_process = ['Referee', 'Assistant 1', 'Assistant 2']
for col in columns_to_process:
    df[col] = df[col].apply(select_last_two_words)
```

```python
# Function to process each value according to the rule
def process_value(value):
    words = value.split()  # Split the string into words
    if len(words) > 2:
        # Keep only the first char of all but the last two words, and keep the last two words as is
        processed_words = [''.join([word[0] for word in words[:-2]])] + words[-2:]
        return ' '.join(processed_words)  # Join the processed words back into a string
    else:
        return value  # Return the value as is if it does not have more than 2 words

# Apply the function to each specified column
columns_to_process = ['Referee', 'Assistant 1', 'Assistant 2']
for col in columns_to_process:
    df[col] = df[col].apply(process_value)

print(df)
```
